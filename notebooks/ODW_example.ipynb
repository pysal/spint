```python
import pysal as ps
from spintW import ODW
```

# With an equal number of origins and destinations (n=16)

```python
origins = ps.weights.lat2W(4, 4)
dests = ps.weights.lat2W(4, 4)
```

```python
origins.n
```

```python
dests.n
```
